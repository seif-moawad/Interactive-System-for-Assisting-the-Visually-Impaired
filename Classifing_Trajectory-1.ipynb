```python
def getPoint(vidPath):
  with mp_holistic.Holistic (static_image_mode=True ,
              min_detection_confidence=0.3,min_tracking_confidence=0.5) as holisitc :
    #right_shoulder=[]
    #left_shoulder=[]
    right_wrist=[]
    left_wrist=[]
    right_elbow=[]
    left_elbow=[]
    left_heel=[]
    right_heel=[]
    left_foot=[]
    right_foot=[]
    points=[]
    
    
    video = cv2.VideoCapture(vidPath)
    while video.isOpened():
        ret, frame = video.read()

        # Recolor Feed
        if ret==True:
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False        

            results = holisitc.process((image))
            if results.pose_landmarks:
                right_hand_landmarks = results.right_hand_landmarks
                left_hand_landmarks = results.left_hand_landmarks

                if right_hand_landmarks:
                    for landmark in right_hand_landmarks.landmark:
                        right_wrist.append(Point(landmark.x, landmark.y, 1))

                if left_hand_landmarks:
                    for landmark in left_hand_landmarks.landmark:
                        left_wrist.append(Point(landmark.x, landmark.y, 2))
                    
#               right_elbow.append(
#                 Point(results.pose_landmarks.landmark[14].x,
#                       results.pose_landmarks.landmark[14].y,
#                       3)
#                 )
#               left_elbow.append(
#                 Point(results.pose_landmarks.landmark[13].x,
#                       results.pose_landmarks.landmark[13].y,
#                       4)
#                )
#               right_heel.append(
#                 Point(results.pose_landmarks.landmark[30].x,
#                       results.pose_landmarks.landmark[30].y,
#                       5)
#                 )
#               left_heel.append(
#                 Point(results.pose_landmarks.landmark[29].x,
#                       results.pose_landmarks.landmark[29].y,
#                       6)
#                )
#               right_foot.append(
#                 Point(results.pose_landmarks.landmark[32].x,
#                       results.pose_landmarks.landmark[32].y,
#                       7)
#                 )
#               left_foot.append(
#                 Point(results.pose_landmarks.landmark[31].x,
#                       results.pose_landmarks.landmark[31].y,
#                       8)
               
               
        else :
          break   
    points = right_wrist + left_wrist + right_elbow + left_elbow + right_heel + left_heel + right_foot + left_foot
    return points   
```

```python
import cv2
import mediapipe as mp
count = 0
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
flag = 0
# Create a VideoCapture object
video = cv2.VideoCapture(1)

with mp_holistic.Holistic(
    static_image_mode=True,
    min_detection_confidence=0.3,
    min_tracking_confidence=0.5
) as holistic:
    image_width = 0
    image_height = 0

    while video.isOpened():
        ret, frame = video.read()

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        if ret == True:
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False

            results = holistic.process(image)

            if results.pose_landmarks:
                right_hand_landmarks = results.right_hand_landmarks
                left_hand_landmarks = results.left_hand_landmarks

                if right_hand_landmarks:
                    for landmark in right_hand_landmarks.landmark:
                        x = int(landmark.x * image.shape[1])
                        y = int(landmark.y * image.shape[0])
                        cv2.circle(frame, (x, y), 5, (0, 255, 0), -1)
                    
                    # Check if required landmarks for drawing a line are present
                    if (
                        right_hand_landmarks.landmark[mp_holistic.HandLandmark.INDEX_FINGER_TIP] and
                        right_hand_landmarks.landmark[mp_holistic.HandLandmark.MIDDLE_FINGER_TIP]
                    ):
                        g = 0
                        if count == 20:
                            cv2.putText(frame, 'Line detected', (20, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
                            # Formatting the data string
                            data_string = f"{'%'}\n"

                            # Path to the output text file
                            output_file = output_file = r"G:\HCI\Project codes\rot.txt"

                            # Saving the data to the text file
                            with open(output_file, "a") as file:
                                file.write(data_string)

                if left_hand_landmarks:
                    for landmark in left_hand_landmarks.landmark:
                        x = int(landmark.x * image.shape[1])
                        y = int(landmark.y * image.shape[0])
                        cv2.circle(frame, (x, y), 5, (0, 255, 0), -1)
                    
                    # Check if required landmarks for drawing a line are present
                    if (
                        left_hand_landmarks.landmark[mp_holistic.HandLandmark.INDEX_FINGER_TIP] and
                        left_hand_landmarks.landmark[mp_holistic.HandLandmark.MIDDLE_FINGER_TIP]
                    ):
                        print('Line line')
                        cv2.putText(frame, 'Line drawn', (20, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
                
            
            cv2.imshow('Holistic Model', frame)
            count = count + 1
            
    video.release()
    cv2.destroyAllWindows()
```
